```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Filtering
img = cv2.imread('albert-einstein.jpg')
kernel = np.ones((5,5), np.float32) / 25

dst = cv2.filter2D(img, -1, kernel)
cv2.imwrite('albert-einstein-filtered.jpg', dst)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dst), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Blur
blur = cv2.blur(img, (5,5))
cv2.imwrite('blur.jpg', blur)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur), plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Gaussian Blur
gblur = cv2.GaussianBlur(img, (5,5), 0)
cv2.imwrite('gblur.jpg', gblur)
plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gblur), plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
plt.subplot(221), plt.imshow(img), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(dst), plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(blur), plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(gblur), plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
