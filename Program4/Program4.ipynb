```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Calculate histogram histogram

img = cv2.imread("image.jpg", 0)
cv2.imshow("Grayscale", img)
cv2.waitKey(0)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.plot(hist)
plt.title("Histogram Image")
plt.xlabel("Pixel Values")
plt.ylabel("Intensity")
plt.show()
```

```python
# Gamma Correction

img = img.astype(np.float32) / 255.0  # Convert image to floating-point in the range [0, 1]

gamma_corrected1 = np.array(255 * (img ** (1/2.2)))  # Gamma correction with gamma = 2.2
gamma_corrected2 = np.array(255 * (img ** (1/1.2)))  # Gamma correction with gamma = 1.2
gamma_corrected3 = np.array(255 * (img ** (1/1.0)))  # Gamma correction with gamma = 1.0 (no correction)
gamma_corrected4 = np.array(255 * (img ** (1/0.5)))  # Gamma correction with gamma = 0.5


cv2.imshow("Gamma 1", gamma_corrected1.astype(np.uint8))
cv2.imshow("Gamma 2", gamma_corrected2.astype(np.uint8))
cv2.imshow("Gamma 3", gamma_corrected3.astype(np.uint8))
cv2.imshow("Gamma 4", gamma_corrected4.astype(np.uint8))

cv2.imwrite("Gamma1.jpg", gamma_corrected1.astype(np.uint8))
cv2.imwrite("Gamma2.jpg", gamma_corrected2.astype(np.uint8))
cv2.imwrite("Gamma3.jpg", gamma_corrected3.astype(np.uint8))
cv2.imwrite("Gamma4.jpg", gamma_corrected4.astype(np.uint8))

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("image.jpg", 0)
equ = cv2.equalizeHist(img)
res = np.hstack((img, equ))

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equ = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Create a 2x2 grid of subplots
plt.subplot(321)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(322)
plt.imshow(equ, cmap='gray')
plt.title("Equalized")

plt.subplot(325)
plt.plot(hist, color='black')
plt.title("Histogram Original")

plt.subplot(326)
plt.plot(hist_equ, color='black')
plt.title("Histogram Equalized")

# Display the figure
plt.show()

cv2.imshow("Histogram Equalized", res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
