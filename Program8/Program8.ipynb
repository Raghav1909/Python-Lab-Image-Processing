```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Erosion
img = cv2.imread('lenna.png', 0)
kernel = np.ones((5,5), np.uint8)
img_erosion = cv2.erode(img, kernel, iterations=1)
cv2.imshow('Input', img)
cv2.imshow('Erosion', img_erosion)
cv2.imwrite('erosion.png', img_erosion)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Dialation
img = cv2.imread('lenna.png', 0)
kernel = np.ones((5,5), np.uint8)
img_dilation = cv2.dilate(img, kernel, iterations=1)
cv2.imshow('Input', img)
cv2.imshow('Dilation', img_dilation)
cv2.imwrite('dilation.png', img_dilation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(221),plt.imshow(img, 'gray'),plt.title('Input'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(img_erosion, 'gray'),plt.title('Erosion'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(img_dilation, 'gray'),plt.title('Dilation'), plt.xticks([]), plt.yticks([])
plt.show()
```
