```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('lenna.png')
img2 = img1.copy()
print(img1.shape)

maskRect = np.zeros(img1.shape[:2],np.uint8)
cv2.rectangle(maskRect,(175,175),(375,375),255,-1)

maskCircle = np.zeros(img2.shape[:2],np.uint8)
cv2.circle(maskCircle,(275,275),100,255,-1)

masked = cv2.bitwise_and(img1,img1,mask = maskRect)
masked2 = cv2.bitwise_and(img2,img2,mask = maskCircle)

cv2.imshow('Original Image',img1)
cv2.imshow('Rectangular Mask', masked)
cv2.imshow('Circular Mask', masked2)

cv2.imwrite('masked.png',masked)
cv2.imwrite('masked2.png',masked2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(221), plt.imshow((cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))), plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow((cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))), plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow((cv2.cvtColor(masked, cv2.COLOR_BGR2RGB))), plt.title('Rectangular Mask'), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow((cv2.cvtColor(masked2, cv2.COLOR_BGR2RGB))), plt.title('Circular Mask'), plt.xticks([]), plt.yticks([])
plt.show()
```
